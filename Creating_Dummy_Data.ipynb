## Creating a customer table

```python
import csv
import random
from faker import Faker
from faker.providers import address
import pandas as pd

# Inisialisasi Faker
fake = Faker('id_ID')
fake.add_provider(address)

# Set seed untuk menghasilkan data yang konsisten
random.seed(42)  # Anda bisa menggunakan angka lain sebagai seed

# Fungsi untuk menghasilkan nomor telepon dengan kode area +62
def generate_phone_number():
    return f"+62{fake.random_number(digits=10)}"  # atau "0{fake.random_number(digits=10)}"

# Fungsi untuk menghasilkan alamat di Yogyakarta
def generate_yogyakarta_address():
    address_parts = fake.street_address().split(',')
    address_parts[0] += ', Yogyakarta'
    return ', '.join(address_parts)

# Menghasilkan data pelanggan dan langsung mengonversinya ke dalam DataFrame
customers_df = pd.DataFrame(
    [[
        i + 1,  # customer_id
        fake.name(),
        generate_yogyakarta_address(),
        generate_phone_number(),
        fake.email(),
        fake.date_of_birth(minimum_age=17, maximum_age=80).strftime('%Y-%m-%d'),
        fake.random_element(elements=('Male', 'Female'))
    ] for i in range(1500)],
    columns=[
        'customer_id', 'name', 'address', 'phone_number', 'email', 'date_of_birth', 'gender'
    ]
)

# # Menyimpan data ke dalam file CSV
# customers_df.to_csv('customers_yogyakarta.csv', index=False, encoding='utf-8')

print("File 'customers_yogyakarta.csv' telah berhasil dibuat.")

# Menampilkan informasi DataFrame
customers_df.info()
```

## Creating Account Table

```python
# Set seed untuk menghasilkan data yang konsisten
random.seed(42)  # Anda bisa menggunakan angka lain sebagai seed

# Fungsi untuk menghasilkan tanggal acak
def random_date(start, end):
    return start + timedelta(days=random.randint(0, (end - start).days))

# Tentukan rentang tanggal
start_date = datetime(2023, 1, 1)
end_date = datetime.now()  # Tanggal saat ini

# Menghasilkan daftar customer_id unik dari 1 hingga 1500
customer_ids = random.sample(range(1, 1501), 1500)

data = []
# Tentukan jumlah akun yang ingin dibuat untuk setiap customer_id
accounts_per_customer = 3  # Misalnya, setiap customer_id memiliki 3 akun

for customer_id in customer_ids:
    for _ in range(accounts_per_customer):
        data.append([
            len(data) + 1,  # account_id (berurutan)
            customer_id,  # customer_id (unik antara 1 dan 1500)
            random.choice(['Savings Account', 'Current Account', 'Fixed Deposit Account']),  # jenis_akun
            round(random.uniform(100000000, 1000000000), 2),  # saldo (acak antara 100.000.000 dan 1.000.000.000)
            random_date(start_date, end_date).strftime('%Y-%m-%d')  # tanggal_pembukaan acak
        ])

with open('accounts.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow([
        'account_id', 'customer_id', 'account_type', 'balance', 'opening_date'
    ])
    writer.writerows(data)

print("File 'accounts.csv' telah berhasil dibuat.")
```

## Creating a transaction table

```python
# Set seed to generate consistent data
random.seed(42)

# Function to generate transaction type
def generate_transaction_type(account_type):
    if account_type == 'Savings Account':
        return random.choice(['Deposit', 'Withdrawal'])
    elif account_type == 'Current Account':
        return random.choice(['Deposit', 'Withdrawal', 'Transfer'])
    elif account_type == 'Fixed Deposit Account':
        return random.choice(['Deposit', 'Withdrawal'])
    return 'Deposit'  # Default

# Function to generate transaction amount
def generate_transaction_amount(account_type):
    if account_type == 'Savings Account':
        return round(random.uniform(100000, 10000000), 2)  # 100,000 - 10,000,000
    elif account_type == 'Current Account':
        return round(random.uniform(50000, 5000000), 2)  # 50,000 - 5,000,000
    elif account_type == 'Fixed Deposit Account':
        return round(random.uniform(1000000, 100000000), 2)  # 1,000,000 - 100,000,000
    return 0.0  # Default

# Function to generate random date and time
def generate_datetime(start_date, end_date):
    time_between_dates = end_date - start_date
    seconds_between_dates = time_between_dates.total_seconds()
    random_number_of_seconds = random.randrange(int(seconds_between_dates))
    random_datetime = start_date + timedelta(seconds=random_number_of_seconds)
    return random_datetime

# Read accounts from CSV file
accounts = []
with open('accounts.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header
    for row in reader:
        account_id = int(row[0])  # Parse account_id as integer
        if 1 <= account_id <= 4500:  # Ensure account_id is within the range 1 to 4500
            accounts.append(row)

transactions = []
transaction_id = 1
start_date = datetime(2023, 1, 1, 0, 0, 0)  # Start from January 1, 2023, at 00:00:00
end_date = datetime(2024, 1, 1, 0, 0, 0)  # Until January 1, 2024, at 00:00:00

# Generate transactions
for i in range(10000):  # 10,000 transactions
    account_index = random.randint(0, len(accounts) - 1)
    account_id = accounts[account_index][0]
    account_type = accounts[account_index][2]
    opening_date = datetime.strptime(accounts[account_index][4], '%Y-%m-%d')  # Opening date

    # Ensure opening_date is not greater than end_date
    if opening_date >= end_date:
        continue  # Skip this iteration if opening date is greater than or equal to the end date

    transaction_type = generate_transaction_type(account_type)
    amount = generate_transaction_amount(account_type)

    # Generate a valid transaction date
    transaction_date = generate_datetime(opening_date, end_date)  # Ensure transaction date >= opening date

    transactions.append([
        transaction_id,
        account_id,
        transaction_type,
        amount,
        transaction_date.strftime("%Y-%m-%d %H:%M:%S")  # Format date
    ])
    transaction_id += 1

# Save transactions to CSV file
with open('transactions_consistent.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow([
        'transaction_id', 'account_id', 'transaction_type', 'amount', 'transaction_date'
    ])
    writer.writerows(transactions)

print("File 'transactions_consistent.csv' has been successfully created.")
```

## Creating a loan table

```python
import csv
import random
from datetime import datetime, timedelta

# Set seed to ensure consistent data generation
random.seed(42)  # You can use a different number as the seed

def generate_loan_amount():
    return round(random.uniform(1000000, 10000000), 2)  # Loan amount between 1,000,000 and 10,000,000

def generate_loan_date(account_opening_date, end_date):
    # Generate a loan date that is later than the account opening date
    time_between_dates = end_date - account_opening_date
    if time_between_dates.days <= 0:
        return None  # Return None if there is no valid date range
    random_number_of_days = random.randrange(1, time_between_dates.days + 1)  # At least 1 day after the opening date
    random_date = account_opening_date + timedelta(days=random_number_of_days)
    return random_date.strftime("%Y-%m-%d")

def generate_loan_term():
    return random.randint(12, 60)  # Loan term between 12 and 60 months

accounts = []
with open('accounts.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header
    for row in reader:
        accounts.append(row)

loans = []
loan_id = 1500  # Start loan_id from 1500
customer_ids = set()  # Ensure each customer has only one loan

end_date = datetime(2024, 1, 1)

for account in accounts:
    customer_id = int(account[1])  # Retrieve customer_id as an integer
    account_opening_date = datetime.strptime(account[4], "%Y-%m-%d")  # Parse account opening date
    if customer_id not in customer_ids and 1 <= customer_id <= 1500:  # Ensure customer_id is within the specified range
        loan_amount = generate_loan_amount()
        loan_date = generate_loan_date(account_opening_date, end_date)  # Generate the loan date
        if loan_date is not None:  # Ensure the loan_date is valid
            loan_term = generate_loan_term()
            loans.append([
                loan_id,
                customer_id,
                loan_amount,
                loan_date,
                loan_term
            ])
            customer_ids.add(customer_id)
            loan_id += 1  # Increment loan_id for the next loan

with open('loans.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow([
        'loan_id', 'customer_id', 'loan_amount', 'loan_date', 'loan_term_months'
    ])
    writer.writerows(loans)

print("The file 'loans.csv' has been successfully created.")
```
